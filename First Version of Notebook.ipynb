# First Version of Notebook: Analyzing COVID-19 Data in Python

COVID-19 is a virus, declared a pandemic by the World Health Organization, that has spread to a number of countries worldwide. 

# Data Source

```python
%matplotlib inline
import pandas
```

# Imported Data From S3 Bucket

```python
df = pandas.read_excel('s3://kropff-ia241/COVID-19-geographic-disbtribution-worldwide-2020-05-01.xlsx')
df
```

# Number of Deaths Per Date

```python
death_per_date= df.groupby('dateRep').count()['deaths']
death_per_date
```

This data set was calculated to find the total number of deaths per day (grouped by date).

```python
death_per_date.plot()
```

The number of deaths per day began increasing at the beginning of March and continues increasing into May.

# Top 20 Countries with the Highest Number of Cases

```python
sum_cases_by_country = df.groupby('countriesAndTerritories').sum()['cases']
sum_cases_by_country.nlargest(20).plot.bar()
```

This calculation finds the top 20 countries with the highest number of total cases.

The United States of America is the country that has the highest number of cases. Spain has the second highest number and Italy is third.

# The Average Number of Cases per Day for the 20 countries with the highest cases

```python
avg_cases_per_date= df.groupby('dateRep').mean()['cases']

avg_cases_per_date.nlargest(20).plot() 
```

This graph calculates the average number of cases per day and it is grouped by the date.

With a data pool of the top 20 countries with the highest number of cases, April 26th had the highest number of average cases.

# China Data Set

```python
china_data = df.loc[df['countriesAndTerritories'] == 'China']
china_data[:20] #the top 15 rows
```

# China Number of Cases per Day

```python
china_data.plot(x='dateRep',y='cases')
```

This calculation finds the total number of cases per day (grouped by date).

The COVID-19 virus is known to have begun in China, and this graph shows that the number of cases per day was the highest mid-February, above 14,000, and then began decreasing.

# Conclusion

This data concludes that the month of April had the highest number of increasing cases and the United States has the highest number of cases in the world. The data from China shows that the number of cases will peak and then begin decreasing, so other countries may experience similar trends.
